### Аналитический подход к регрессии

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression 

X, y = make_regression(300, 1, random_state=42, bias=100., noise= 8., shuffle=True)
plt.scatter(X, y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3)

lr = LinearRegression()
lr.fit(Xtrain, ytrain)

pred_lr = lr.predict(Xtest)
```

```python
plt.scatter(Xtest, ytest, c='blue')
plt.plot(Xtest, pred_lr, c="green")
```

```python
from sklearn.metrics import r2_score

error = r2_score(ytest, pred_lr)
error
```

### Численный подход к регрессии

```python
from sklearn.linear_model import SGDRegressor

sgd = SGDRegressor()
sgd.fit(Xtrain, ytrain)

sgd_pred = sgd.predict(Xtest)
sgd_error = r2_score(ytest, sgd_pred)
sgd_error
```

```python
plt.scatter(Xtest, ytest, c='blue')
plt.plot(Xtest, sgd_pred, c='green')
```

### Вывод
Сравнивая значения ошибок SGDRegressor и LinearRegressor сделаем вывод, что из-за малого различия в R2 метрике, разницы в качестве нет.
